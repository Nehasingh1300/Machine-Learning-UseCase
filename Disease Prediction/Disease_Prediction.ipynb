```python
# importing the library
import pandas as pd
```

```python
# Reading Testing Dataset
df = pd.read_csv("training_data.csv")
```

```python
# Viewing Dataset
df.head()
```

```python
# Checking shape of Dataset
df.shape
```

```python
# Storing prognosis(prediction column) in y_train dataframe
y_train =df["prognosis"]

y_train.head(50)
```

```python
# deleting prediction column as we have stored in y_train
del df["prognosis"]

# Unnamed column as it is of no use to us.
del df["Unnamed: 133"]
```

```python
df.head()
```

```python
# Checking the NULL Values
df.isnull().sum()
```

As we can see our dataset has no NULL values, and even we can not do any Data Cleaning, as the dataset is not that good. So, we can go on with this or else we can use a better dataset.

```python
# Storing training dataset in X
X = df

# Stroing prediction column in Y
Y = y_train
```

Now we will divide the dataset into train and test for doing the dataset.

---



```python
# importing sklearn library for train test slpitting
from sklearn.model_selection import train_test_split
X_train, X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.1,stratify=Y,random_state=2)
```

```python
# Checking the shape of test train dataset
print(X.shape, X_train.shape, X_test.shape)
```

```python
# importing the libraries
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn import metrics

# building model using Logistic Regression
model_l = LogisticRegression()

# Fitting the model with training dataset
model_l.fit(X_train,Y_train)

# Prdicting the column using the model
lr_prediction = model_l.predict(X_test)

# Checking the accuracy of model
print("Accuracy: ", metrics.accuracy_score(lr_prediction,Y_test))
```

Accuracy came out to be 100%(1.0) as we can see that dataset if not big or complicated enough. 

Even if the prognosis column depends upon one column then also it will give the right value. 

So, there is no issue with the method use but with the dataset.
